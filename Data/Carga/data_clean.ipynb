**Procedemos a transformar los dataframes para poder proceder a la carga**

Establecemos un codigo de plataforma que permita crear una relacion a una tabla plataforma para saber idenficar a que servicio pertenece cada titulo 

```python

df_netflix['plataforma_id'] = 1
df_disney['plataforma_id'] = 2
df_netflix.insert(0, "id", np.nan) 
df_disney.insert(0, "id", np.nan) 
```

**Porcedemos a transformar las fechas para posteriormente poder relacionarlo a una tabla calendario**

```python
df_netflix['date_added'] = pd.to_datetime(df_netflix['date_added'], errors='coerce')
df_disney['date_added'] = pd.to_datetime(df_disney['date_added'], errors='coerce')
df_netflix['release_year'] = pd.to_datetime(df_netflix['release_year'], errors='coerce')
df_disney['release_year'] = pd.to_datetime(df_disney['release_year'], errors='coerce', format='%Y')
df_netflix['release_year'] = df_netflix['release_year'].dt.strftime('%Y')
df_disney['release_year'] = df_disney['release_year'].dt.strftime('%Y')
# df_disney['release_year'].head(10)
```

**IMPORTAMOS A LA BASE DE DATOS LA INFORMACION PROCESADA**

```python
from sqlalchemy.types import Integer
from sqlalchemy.types import DATE
from sqlalchemy.types import VARCHAR 
from sqlalchemy.types import INTEGER
from sqlalchemy.types import Text


```

```python
df_disney.to_sql('shows', con=engine, if_exists='append', index=False,
           dtype={"show_id": VARCHAR(10), "type":VARCHAR(15), "title":VARCHAR(100), "director":VARCHAR(80),
            "cast":VARCHAR(200), "country":VARCHAR(50), "date_added": DATE, "release_year":INT, "rating":VARCHAR(15),
            "duration":VARCHAR(50), "listed_in":VARCHAR(150), "description":VARCHAR(250),"plataforma_id":Integer})
```

```python
df_db = pd.read_sql('SELECT * FROM shows', con=engine)
df_db.dtypes
# df_db.to_sql()
```

```python

try:
    df_netflix.to_sql('shows', con=engine, if_exists='replace', index=False)
    print('Carga Exitosa')
except Exception as ex:
    print('Error al cargar')
    print(ex)
```
